```python
import pandas as pd

# Specify the path to your Excel file
file_path = 'nitrilase_descriptors.xlsx'

# Read the Excel file
df = pd.read_excel(file_path)
df
```

```python
df = df.loc[:, ["protein", "nitriles", "ZINC ID", "activity"]]
df
```

```python
from Bio import SeqIO

sequences = []

# Read the FASTA file
for record in SeqIO.parse("all_sequences_Mou.fasta", "fasta"):
    id_ = record.id
    sequence = str(record.seq)

    sequences.append(sequence)
```

```python
import pandas as pd

compounds = pd.read_csv("nitrilases_compounds_smiles.csv", sep=" ")
compounds = compounds.loc[:, ["smiles", "source_label"]]
compounds.columns = ["SMILES", "ZINC ID"]
```

```python
df["Enzyme_sequence"] = sequences
```

```python
df
```

```python
pd.merge(df, compounds, on="ZINC ID", how="left").drop_duplicates().to_csv("nitrilases_dataset.csv", index = False)
```

```python
import pandas as pd

nitrilases = pd.read_csv("nitrilases_dataset.csv")
nitrilases
```

```python
nitrilases.columns = ["Enzyme ID", "Substrate Name", "Substrate ID", "activity", "Sequence", "SMILES"]
nitrilases["Binding"] = [int(y > 2) for y in nitrilases["activity"]]
nitrilases
```

```python
nitrilases["Publication"] = ["https://doi.org/10.1002/prot.26019"] * nitrilases.shape[0]
```

```python
nitrilases.to_csv("mou_data.csv", index=False)
```
