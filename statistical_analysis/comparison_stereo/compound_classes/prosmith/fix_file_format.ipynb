```python
import os

import pandas as pd


# Read the input file
for similarity in [20, 40, 60, 80]:
    df = pd.read_csv(f'class_results_{similarity}.csv')

    # Group by class_name
    for class_name, group in df.groupby('class_name'):
        # Replace class_name with similarity
        group = group.drop(columns=['class_name'])
        group['similarity'] = similarity

        # Save to a new file
        output_filename = f'results_binding_np_classifier_compounds_{class_name}.csv'
        if os.path.exists(output_filename):
            output = pd.read_csv(output_filename)
            output = pd.concat((output, group))
            output.to_csv(output_filename, index=False)
        else:

            group.to_csv(output_filename, index=False)
        print(f'Saved: {output_filename}')
```
