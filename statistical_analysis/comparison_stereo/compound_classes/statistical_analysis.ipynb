```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from math import pi

def generate_metric_radar_plot(models, compound_files, similarity, metric='mcc', exceptions=[]):
    name_mappings = {
        "Polyketides": "Polyketides",
        "Fatty_acids": "Fatty acids",
        "Carbohydrates": "Carbohydrates",
        "Shikimates_Phenylpropanoids": "Shikimates and Phenylpropanoids",
        "Terpenoids": "Terpenoids",
        "Amino_acids_Peptides": "Amino acids and Peptides",
        "Alkaloids": "Alkaloids"
    }

    all_mcc_values = {model: {} for model in models}

    for exception in exceptions:
        del compound_files[exception]

    for model in models:
        for compound, file in compound_files.items():
            try:
                df = pd.read_csv(os.path.join(model, file))
                df = df[df["similarity"] == similarity]
                all_mcc_values[model][compound] = df[metric].mean()
            except Exception as e:
                print(f"Erro ao processar {model}/{compound}: {e}")

    categories = list(compound_files.keys())
    categories = [name_mappings[category] for category in categories]
    N = len(categories)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    models_map = {
        "esp": "ESP (State-of-the-art)",
        "esm2_3b": "XGBoost-NPClassifierFP-ESM2",
        "prosmith": "ProSmith"
    }

    # Define colorblind-friendly colors for each model
    model_colors = {
        "esp": "#1f77b4",          # Blue
        "prosmith": "#2ca02c",     # Green
        "esm2_3b": "#ff7f0e",      # Orange
    }

    # Create the radar chart
    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'polar': True})

    for i, model in enumerate(all_mcc_values):
        values = list(all_mcc_values[model].values())
        values += values[:1]
        ax.plot(angles, values, linewidth=2, label=models_map[model], color=model_colors[model])
        ax.fill(angles, values, alpha=0.1, color=model_colors[model])

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories)
    ax.set_title(f'Model comparison of MCC based on compound type for similarity threshold {similarity}%', size=16, y=1.1)

    # Save the radar chart without legend
    plt.tight_layout()
    plt.savefig(f'compound_mcc_radar_chart_all_models_{similarity}.png', dpi=200, bbox_inches='tight', transparent=True)
    plt.close()

    # Create a separate figure for the legend
    fig_legend = plt.figure(figsize=(6, 1))
    handles, labels = ax.get_legend_handles_labels()
    fig_legend.legend(handles, labels, loc='center', ncol=len(models))
    fig_legend.savefig(f'compound_mcc_radar_chart_legend_{similarity}.png', dpi=200, bbox_inches='tight', transparent=True)
    plt.close()
```

```python
models = ["esm2_3b", "esp", "prosmith"]

# Dicionário de ficheiros de compostos
compound_files = {
    "Polyketides": "results_binding_np_classifier_compounds_Polyketides.csv",
    "Fatty_acids": "results_binding_np_classifier_compounds_Fatty acids.csv",
    "Carbohydrates": "results_binding_np_classifier_compounds_Carbohydrates.csv",
    "Shikimates_Phenylpropanoids": "results_binding_np_classifier_compounds_Shikimates and Phenylpropanoids.csv",
    "Terpenoids": "results_binding_np_classifier_compounds_Terpenoids.csv",
    "Amino_acids_Peptides": "results_binding_np_classifier_compounds_Amino acids and Peptides.csv",
    "Alkaloids": "results_binding_np_classifier_compounds_Alkaloids.csv"
}

generate_metric_radar_plot(models, compound_files, 60, metric='mcc')
```
