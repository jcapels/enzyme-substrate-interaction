```python
import pandas as pd

kroll_data = pd.read_csv("merged_dataset_with_representation.csv")
kroll_data
```

```python
kroll_data = kroll_data[kroll_data["Binding"]==1]
kroll_data.columns = ["Enzyme ID", "Substrate ID", "Evidence", "Binding", "SMILES", "Sequence"]
kroll_data["Publication"] = ["https://doi.org/10.1038/s41467-023-38347-2"] * kroll_data.shape[0]
```

```python
kroll_data.to_csv("kroll_data.csv", index=False)
```

```python
import pandas as pd

train_dataset = pd.read_csv("train_dataset_w_representation_filtered.csv")

# split dataset into train and test
from sklearn.model_selection import train_test_split

train_test_splits = train_test_split(train_dataset, test_size=0.2, random_state=42, stratify=train_dataset["Binding"])
train_dataset = train_test_splits[0]
validation_dataset = train_test_splits[1]
```

```python
train_dataset.Binding.value_counts(), validation_dataset.Binding.value_counts()
```

```python
# write train and validation datasets to csv
train_dataset.to_csv("train_dataset_w_split.csv", index=False)
validation_dataset.to_csv("validation_dataset.csv", index=False)
```
