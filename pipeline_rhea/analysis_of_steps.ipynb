```python
import pandas as pd

dataset_raw_rhea = pd.read_csv("RHEA_enzyme_compound_pairs_sequence_smiles_filtered.csv")
dataset_raw_rhea
```

```python
uniprot = pd.read_csv("swiss_prot_enzymes.csv")
uniprot
```

```python
enzymes = uniprot[~uniprot["EC"].isna()]
enzymes.columns = ["uniprot_id", "name", "sequence", "EC number"]
enzymes.drop(["name", "sequence"], inplace=True, axis=1)
```

```python
enzymes
```

```python
dataset = dataset_raw_rhea[dataset_raw_rhea["EC number"].isna()].drop(["EC number"], axis=1)
dataset = pd.merge(dataset, enzymes, on="uniprot_id", how="inner")
dataset
```

```python
incomplete_ecs = dataset[(~dataset["EC number"].str.contains(";") & dataset["EC number"].str.contains("-"))]
dataset_raw_rhea = dataset_raw_rhea[~dataset_raw_rhea["EC number"].isna()]
dataset_raw_rhea = pd.concat((dataset_raw_rhea, incomplete_ecs))
```
