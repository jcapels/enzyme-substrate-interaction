```python
import pandas as pd

df_RHEA_smiles_sequences = pd.read_csv("RHEA_enzyme_compound_pairs_sequence_smiles.csv")
print(df_RHEA_smiles_sequences.shape)
print(len(df_RHEA_smiles_sequences["uniprot_id"].unique()))
print(len(df_RHEA_smiles_sequences["CHEBI_ID"].unique()))
```

```python
df_RHEA_smiles_sequences
```

```python
cofactors = pd.read_csv("cofactors.tsv", sep="\t", )
```

```python
cofactors = cofactors.index.values
```

```python
df_RHEA_smiles_sequences_without_cofactors =  df_RHEA_smiles_sequences[~df_RHEA_smiles_sequences["CHEBI_ID"].isin(cofactors)]
```

```python
df_RHEA_smiles_sequences_without_cofactors
```

```python
print(df_RHEA_smiles_sequences_without_cofactors.shape)
print(len(df_RHEA_smiles_sequences_without_cofactors["uniprot_id"].unique()))
print(len(df_RHEA_smiles_sequences_without_cofactors["CHEBI_ID"].unique()))
```

```python
len(df_RHEA_smiles_sequences_without_cofactors["CHEBI_ID"].unique())
```

```python
names = pd.read_csv("names.tsv", sep="\t")
names["COMPOUND_ID"] = [f"CHEBI:{ID}" for ID in names["COMPOUND_ID"]]
names
```

```python
# convert the values for ID and NAME in the dataframe to dictionary
names_dict = names.set_index('COMPOUND_ID')['NAME'].to_dict()
names_dict
```

```python
# make plot of the most common compounds
import matplotlib.pyplot as plt

twenty_compounds_more_used = df_RHEA_smiles_sequences_without_cofactors["CHEBI_ID"].value_counts()[:200].to_dict()
new_dict = {}
for key in twenty_compounds_more_used.keys():
    if key in names_dict:
        new_dict[names_dict[key]] = twenty_compounds_more_used[key]
    else:
        new_dict[key] = twenty_compounds_more_used[key]

new_dict
```

```python
compounds_to_remove = pd.read_csv("300_compounds_more_used.csv", sep="\t")
compounds_to_remove = compounds_to_remove[compounds_to_remove["remove"]=="yes"]["CHEBI_ID"].values
compounds_to_remove
```

```python
df_RHEA_smiles_sequences_without_cofactors = df_RHEA_smiles_sequences_without_cofactors[~df_RHEA_smiles_sequences_without_cofactors["CHEBI_ID"].isin(compounds_to_remove)]
```

```python
# make plot of the most common compounds
import matplotlib.pyplot as plt

twenty_compounds_more_used = df_RHEA_smiles_sequences_without_cofactors["CHEBI_ID"].value_counts()[:20].to_dict()
new_dict = {}
for key in twenty_compounds_more_used.keys():
    if key in names_dict:
        new_dict[names_dict[key]] = twenty_compounds_more_used[key]
    else:
        new_dict[key] = twenty_compounds_more_used[key]

new_dict
```

```python
# convert dictionary to plot 
plt.bar(new_dict.keys(), new_dict.values())
plt.xticks(rotation=90)
```

```python
df_RHEA_smiles_sequences_without_cofactors.to_csv("RHEA_enzyme_compound_pairs_sequence_smiles_without_cofactors.csv", index=False)
```
