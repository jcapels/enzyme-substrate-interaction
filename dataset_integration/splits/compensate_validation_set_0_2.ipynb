```python
import pandas as pd

curated_dataset = pd.read_csv("curated_dataset.csv")
assignments = pd.read_csv("graphpart_assignments_0_2_train_test_split_curated.csv")
assignments
```

```python
train_dataset_ids = assignments[assignments["cluster"]==0]["AC"]
validation_dataset_ids = assignments[assignments["cluster"]==2]["AC"]
test_dataset_ids = assignments[assignments["cluster"]==1]["AC"]
```

```python
train_dataset = curated_dataset[curated_dataset["Enzyme ID"].isin(train_dataset_ids)]
train_dataset
```

```python
proteins_to_add = train_dataset[train_dataset["Binding"] == 0].drop_duplicates(subset=["Enzyme ID"]).loc[:, ["Enzyme ID"]].sample(frac=0.1)
```

```python
from plants_sm.io.pickle import read_pickle, write_pickle

splits = read_pickle("splits_0_2_proteins_train_val_test.pkl")
```

```python
proteins_to_add
```

```python
train_ids = splits[0][0]
train_ids = train_ids[~train_ids.isin(proteins_to_add["Enzyme ID"])]
```

```python
validation_ids = splits[0][1]
validation_ids = pd.concat([validation_ids, proteins_to_add["Enzyme ID"]])
test_ids = splits[0][2]
```

```python
write_pickle(
    
    "splits_0_2_proteins_train_val_test_curated_changed.pkl",
    [
        (train_ids, validation_ids, test_ids),
    ]
)
```
