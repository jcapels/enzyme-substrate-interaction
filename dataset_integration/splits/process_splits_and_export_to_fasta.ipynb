```python
import pandas as pd
from plants_sm.io.pickle import read_pickle

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

def write_fasta(df, output_file):
    records = []
    for _, row in df.iterrows():
        seq_record = SeqRecord(
            Seq(row["Sequence"]),
            id=row["Enzyme ID"],
            description=""
        )
        records.append(seq_record)
    SeqIO.write(records, output_file, "fasta")

def get_sequences_from_splits(threshold):

    splits = read_pickle(f"splits_{threshold}_proteins_train_val_test.pkl")
    _, valid_ids, test_ids = splits[0]
    print(len(valid_ids), len(test_ids))
    dataset = pd.read_csv("curated_dataset.csv")
    dataset = dataset[dataset["Enzyme ID"].isin(valid_ids) | dataset["Enzyme ID"].isin(test_ids)]
    dataset.drop_duplicates(subset=["Enzyme ID"], inplace=True)

    print(dataset.shape)
    write_fasta(dataset, f"curated_dataset_valid_test_{threshold}.fasta")


get_sequences_from_splits("0_8")
get_sequences_from_splits("0_6")
get_sequences_from_splits("0_4")
get_sequences_from_splits("0_2")
```

```python
dataset = pd.read_csv("augmented_dataset_descriptors_available.csv")
dataset.drop_duplicates(subset=["Enzyme ID"], inplace=True)
write_fasta(dataset, f"augmented_dataset.fasta")
```
