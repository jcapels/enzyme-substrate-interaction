```python
import pandas as pd

overlapping_range = pd.read_csv("overlapping_range.csv")
overlapping_range
```

```python
dataset = pd.read_csv("pdb_chemical_components_curated_dataset.csv")
dataset.replace({"PDB_ID": "pdb"}, inplace=True)
dataset
```

```python
# rename column 'PDB_ID' to 'pdb' (if present) and show the result
dataset.rename(columns={"PDB_ID": "pdb"}, inplace=True)

dataset = dataset.iloc[:, :2]
```

```python
overlapping_range = pd.merge(overlapping_range, dataset, on="pdb", how="left")
overlapping_range.drop_duplicates(subset=["pdb", "compound"], inplace=True)
overlapping_range
```

```python
from plants_sm.io.pickle import write_pickle

pairs_to_select = list(zip(overlapping_range['Uniprot_ID'], overlapping_range['compound']))

write_pickle("pairs_to_select.pkl", pairs_to_select)
```

```python
dataset = pd.read_csv("../curated_dataset_no_stereochemistry_duplicates.csv")
```

```python
mask = dataset.apply(lambda row: (row['Enzyme ID'], row['Substrate ID']) in pairs_to_select, axis=1)

# Select only the matching rows
dataset_selected = dataset[mask]
dataset_selected
```
