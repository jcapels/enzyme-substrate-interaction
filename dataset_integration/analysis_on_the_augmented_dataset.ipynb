```python
import pandas as pd

dataset = pd.read_csv('augmented_dataset_descriptors_available.csv')
dataset
```

```python
rhea_negatives = rhea[rhea.interaction == 0].loc[:, ["SMILES", "uniprot_id", "CHEBI_ID", "sequence", "interaction"]]
rhea_negatives.columns = ["SMILES", "Enzyme ID", "Substrate ID", "Sequence", "Binding"]
rhea_negatives = rhea_negatives[rhea_negatives["Enzyme ID"].isin(dataset.loc[:, "Enzyme ID"])]
```

```python
dataset = pd.concat([dataset, rhea_negatives], ignore_index=True)
dataset
```

```python
dataset.to_csv('augmented_dataset_with_rhea_negatives.csv', index=False)
```
