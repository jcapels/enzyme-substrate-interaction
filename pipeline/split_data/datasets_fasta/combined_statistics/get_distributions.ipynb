```python
import pandas as pd

train_test_40 = pd.read_csv("train_vs_test_40_combined.csv")
train_val_40 = pd.read_csv("train_vs_val_40_combined.csv")
val_test_40 = pd.read_csv("val_vs_test_40_combined.csv")
```

```python
max_identity_per_protein_train_test_40 = train_test_40.loc[train_test_40.groupby('protein_1')['identity'].idxmax()]
max_identity_per_protein_train_val_40 = train_val_40.loc[train_val_40.groupby('protein_1')['identity'].idxmax()]
max_identity_per_protein_val_test_40 = val_test_40.loc[val_test_40.groupby('protein_1')['identity'].idxmax()]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Combine the data into a single DataFrame for plotting
data = pd.concat([
    max_identity_per_protein_train_test_40.assign(dataset='Train vs Test'),
    max_identity_per_protein_train_val_40.assign(dataset='Train vs Val'),
    max_identity_per_protein_val_test_40.assign(dataset='Val vs Test')
])

# Set up the plot
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# Create a boxplot
ax = sns.boxplot(x='dataset', y='identity', data=data, palette='Set2')

# Overlay the data points as a swarm plot
sns.stripplot(x='dataset', y='identity', data=data, color='black', alpha=0.5, jitter=True, size=3)

ax.set_ylim((0, 1))

# Add labels and title
ax.set_xlabel('Dataset')
ax.set_ylabel('Identity')
ax.set_title('Distribution of Maximum Identity Values by Dataset - 40% threshold')

# Show the plot
plt.tight_layout()
plt.savefig('boxplots_with_points_40.png', format='PNG', dpi=200, bbox_inches='tight', transparent=True)
plt.show()
```
