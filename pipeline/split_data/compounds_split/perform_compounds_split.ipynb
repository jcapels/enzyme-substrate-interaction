```python
import pandas as pd

curated_dataset = pd.read_csv("../curated_dataset.csv", index_col=0)
```

```python
curated_dataset.drop_duplicates(subset=["Substrate ID"], inplace=True)
unique_compounds = curated_dataset.loc[:, ["SMILES", "Substrate ID"]]
```

```python
from deepmol.datasets import SmilesDataset

dataset = SmilesDataset(smiles=unique_compounds["SMILES"].tolist(),
              ids=unique_compounds["Substrate ID"].tolist())
```

```python
from deepmol.splitters import GraphPartSimilaritySplitter

splitter = GraphPartSimilaritySplitter(
    threshold=0.8,
    stratified=True,
)
train, valid, test = splitter.train_valid_test_split(dataset, frac_train=0.7, frac_valid=0.1, frac_test=0.2)
valid_set_08 = set(valid.ids)
test_set_08 = set(test.ids)
train_set_08 = set(train.ids)

# export to pickle file
import pickle
with open("splits_compounds_08.pkl", "wb") as f:
    pickle.dump([(train_set_08, valid_set_08, test_set_08)], f)
```
