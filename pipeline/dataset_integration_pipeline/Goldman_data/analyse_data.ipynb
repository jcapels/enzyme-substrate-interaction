```python
import pandas as pd


phospatase = pd.read_csv("phosphatase_chiral_binary.csv")
phospatase.columns = ["Index", "Sequence", "SMILES", "Binding"]
```

```python
import pandas as pd

phospatase_sequences = pd.read_csv("phosphatase_metadata/phosphatase_sequence_file.txt", sep="\t")
phospatase_sequences = phospatase_sequences.loc[:, ["Entry", "Sequence"]]
```

```python
phospatase.shape
```

```python
import numpy as np


phospatase = pd.merge(phospatase, phospatase_sequences, on="Sequence")
phospatase.columns = ["Index", "Sequence", "SMILES", "Binding", "Enzyme ID"]
# Check for None or NaN values
nan_mask = np.array([str(x).lower() == 'nan' for x in np.array(phospatase.loc[:, "Enzyme ID"])])

# Count the number of NaN values
nan_count = np.sum(nan_mask)
nan_count
```

```python

def convert_dataset(phospatase, name):

    # Sample data: assuming phosphatase.Sequence.unique() returns a list of unique sequences
    unique_sequences = phospatase.SMILES.unique()

    # Create a dictionary with keys labeled as "phosphatase 1", "phosphatase 2", etc.
    phosphatase_compounds_dict = {seq: f"compound_{name}_dataset {i+1}" for i, seq in enumerate(unique_sequences)}

    compound_ids = []

    for i, row in phospatase.iterrows():
        smiles = row["SMILES"]
        compound_ids.append(phosphatase_compounds_dict[smiles])

    phospatase["Substrate ID"] = compound_ids
    return phospatase

```

```python
phospatase = convert_dataset(phospatase, "phospatase")
phospatase
```

```python
phospatase.to_csv("phospatase_data.csv", index=False)
```

# DUF 

```python
import pandas as pd


duf = pd.read_csv("duf_binary.csv")
duf.columns = ["Index", "Sequence", "SMILES", "Binding"]
duf
```

```python
from itertools import groupby
from typing import Iterable, Tuple


def fasta_iter(fasta_name: str) -> Iterable[Tuple[str, str]]:
    """fasta_iter.

     modified from Brent Pedersen
     Correct Way To Parse A Fasta File In Python
     given a fasta file. yield tuples of header, sequence

    Args:
        fasta_name (str): fasta_name

    Returns:
        Iterable[tuple] of (headerStr, seq)
    """
    # open fasta file
    fh = open(fasta_name)

    # ditch the boolean (x[0]) and just keep the header or sequence since
    # we know they alternate.
    faiter = (x[1] for x in groupby(fh, lambda line: line[0] == ">"))

    for header in faiter:
        # drop the ">"
        headerStr = header.__next__()[1:].strip()

        # join all sequence lines to one.
        seq = "".join(s.strip() for s in faiter.__next__())

        yield (headerStr, seq)

seq_mapping = {
        seq.replace("-", ""): name.replace("BKACE", "KCE")
        for name, seq in fasta_iter("duf_metadata/duf_msa.txt")
    }
```

```python
sequences_ids = []

for i, row in duf.iterrows():
    sequence = row["Sequence"]
    sequences_ids.append(seq_mapping[sequence])
duf["Enzyme ID"] = sequences_ids
duf
```

```python
duf = convert_dataset(duf, "duf")
duf.to_csv("duf_data.csv", index=False)
```
