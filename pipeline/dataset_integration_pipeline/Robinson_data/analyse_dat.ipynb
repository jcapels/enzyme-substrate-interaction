```python
from Bio import SeqIO

sequences = {}

# Read the FASTA file
for record in SeqIO.parse("73_OleA_JGI_unaligned.fasta", "fasta"):
    id_ = record.id
    if "Pseudoxanthomonas" in id_:
        id_ = "Pseudoxanthomonas_sp."
    elif "Leifsonia" in id_:
        id_ = "Leifsonia_sp."
    sequence = str(record.seq)

    sequences[id_] = sequence
```

```python
import pandas as pd
import numpy as np

robinson_data = pd.read_csv("robinson_data.csv")
enzymes = []
enzyme_sequences = []
compounds = []
bindings = []

for i, data in enumerate(robinson_data["id"].str.split("_")):
    compounds.append(data[1].replace(".", " "))
    enzymes.append(data[0])
    enzyme = data[0].replace(" ", "_")
    found = False
    for sequence in sequences:
        if enzyme in sequence:
            enzyme_sequences.append(sequences[sequence])
            found = True
    
    if not found:
        print(enzyme)


    binding = robinson_data.loc[i, "is_active"]
    if binding == "Y":
        binding = 1
    else:
        binding = 0
    
    bindings.append(binding)
```

```python
len(enzymes), len(compounds), len(enzyme_sequences), len(bindings)
```

```python
import pandas as pd

new_dataset = pd.DataFrame(
    {"Enzyme": enzymes,
     "Compound": compounds,
     "Enzyme_sequence": enzyme_sequences,
     "Binding": bindings}
)
new_dataset
```

```python
new_dataset["Compound"] = new_dataset.Compound.str.replace("oxadiazole", "oxidazole")
```
