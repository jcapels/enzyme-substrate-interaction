```python
import pandas as pd

compounds_predicted_pathways = pd.read_csv('compounds_predicted_pathways.csv')
compounds_predicted_pathways
```

```python
from plants_sm.io.pickle import read_pickle, write_pickle

def create_pathway_dict(pickle_file, compounds_predicted_pathways, output_file):
    _, _, test_dataset = read_pickle(pickle_file)[0]
    compounds_predicted_pathways_test = compounds_predicted_pathways[compounds_predicted_pathways["ID"].isin(test_dataset)]

    pathway_to_compounds = (
        compounds_predicted_pathways_test
        .assign(Pathways=compounds_predicted_pathways_test.Pathways.str.split(','))
        .explode('Pathways')
        .assign(Pathways=lambda x: x.Pathways.str.strip())
    )

    # Create a dictionary: {pathway: list of compound IDs}
    pathway_dict = (
        pathway_to_compounds
        .groupby('Pathways')['ID']
        .apply(list)
        .to_dict()
    )
    write_pickle(output_file, pathway_dict)

create_pathway_dict(
    'splits_compounds_02.pkl',
    compounds_predicted_pathways,
    'pathway_to_compounds_split_02.pkl'
)

create_pathway_dict(
    'splits_compounds_04.pkl',
    compounds_predicted_pathways,
    'pathway_to_compounds_split_04.pkl'
)

create_pathway_dict(
    'splits_compounds_06.pkl',
    compounds_predicted_pathways,
    'pathway_to_compounds_split_06.pkl'
)

create_pathway_dict(
    'splits_compounds_08.pkl',
    compounds_predicted_pathways,
    'pathway_to_compounds_split_08.pkl'
)
```
