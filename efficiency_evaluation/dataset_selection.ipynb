```python
import pandas as pd 

dataset = pd.read_csv("../dataset_integration/curated_dataset.csv")
```

```python
import numpy as np

unique_substrates = np.unique(dataset["Substrate ID"])
unique_enzymes = np.unique(dataset["Enzyme ID"])
```

```python
unique_substrates_100 = unique_substrates[:100]
unique_enzymes_100 = unique_enzymes[:100]

unique_substrates_300 = unique_substrates[:300]
unique_enzymes_1000 = unique_enzymes[:1000]

unique_substrates_700 = unique_substrates[:700]
unique_enzymes_5000 = unique_enzymes[:5000]

unique_enzymes_7000 = unique_enzymes[:7000]
```

```python
dataset_100_100 = dataset[dataset["Substrate ID"].isin(unique_substrates_100) & dataset["Enzyme ID"].isin(unique_enzymes_100)]
unique_substrates_dataset_100_100 = np.unique(dataset_100_100["Substrate ID"])
print(len(unique_substrates_dataset_100_100))
unique_enzymes_dataset_100_100 = np.unique(dataset_100_100["Enzyme ID"])
print(len(unique_enzymes_dataset_100_100))
print(dataset_100_100.shape)
dataset_100_100.to_csv("dataset_100_100.csv", index=False)
```

```python
dataset_300_1000 = dataset[dataset["Substrate ID"].isin(unique_substrates_300) & dataset["Enzyme ID"].isin(unique_enzymes_1000)]
unique_substrates_dataset_300_1000 = np.unique(dataset_300_1000["Substrate ID"])
print(len(unique_substrates_dataset_300_1000))
unique_enzymes_dataset_300_1000 = np.unique(dataset_300_1000["Enzyme ID"])
print(len(unique_enzymes_dataset_300_1000))
print(dataset_300_1000.shape)
dataset_300_1000.to_csv("dataset_300_1000.csv", index=False)
```
